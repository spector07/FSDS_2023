```python
# 1. What is the primary goal of Object-Oriented Programming (OOP)?

OOP is used to obtain the code reusablity which helps to avoid redunduncy.
Scalable and maintainable code can be written with OOP.
It allows to model the real world concepts as objects, their properties as attributes and behaviour as methods.
```

```python
# 2. What is an object in Python?

An object is an instance of a class. It enables to access the class properties and methods.
```

```python
# 3. What is a class in Python?

Class is blueprint. It consists various attributes, constructor to intialize objects and methods to define as behaviors
of perticular class.
```

```python
# 4. What are attributes and methods in a class?

Attributes are the properties of class. Any variable declared in class is called attribute of that class.
Methods are behaviors of the class. Any function defined in class is called method of that class.
```

```python
# 5. What is the difference between class variables and instance variables in Python?

Class variables are defined within the class, instance variables are defined within class methods.
On changing the values of class variables affects all instances, this does not happen with instance variables
Class variables are used to store values which will be shared among all instances of class, such as a constant value.
Instance variables are used to store values which is unique for each instance of class.


For eg.

class Test:
    a = 'xyz' # this is class variable
    
    def __init__(self, atr):
        self.atr = atr  # this is instance variable

        
```

```python
# 6. What is the purpose of the self parameter in Python class methods?

self parameter refers to the current instance of the class.
Python does not have @ annotation instead it uses self parameter as the first parameter in class methods.
It is not neccessary to write self but we can call it anything, but it should be the first parameter of the class methods.
```

```python
# For a library management system, you have to design the "Book" class with OOP
# principles in mind. The “Book” class will have following attributes:
# a. title: Represents the title of the book.
# b. author: Represents the author(s) of the book.
# c. isbn: Represents the ISBN (International Standard Book Number) of the book.
# d. publication_year: Represents the year of publication of the book.
# e. available_copies: Represents the number of copies available for checkout.
# The class will also include the following methods:
# a. check_out(self): Decrements the available copies by one if there are copies
# available for checkout.
# b. return_book(self): Increments the available copies by one when a book is
# returned.
# c. display_book_info(self): Displays the information about the book, including its
# attributes and the number of available copies.


import logging

logging.basicConfig(filename="Book.log", format='%(levelname)s : %(message)s', level=logging.INFO)


class Book:

    def __init__(self, title, author, isbn, publication_year, available_copies):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.publication_year = publication_year
        self.available_copies = available_copies

    def check_out(self):
        try:
            if self.available_copies > 0:
                self.available_copies = self.available_copies - 1
                logging.info(f'{self.title} checked out, {self.available_copies} available copies')
                return f'{self.title} - {self.available_copies} available copies'
            else:
                return f'Sufficient copies of {self.title} not available'
        except Exception as e:
            logging.error(e)

    def return_book(self):
        try:
            self.available_copies = self.available_copies + 1
            logging.info(f'{self.title} is returned, {self.available_copies} available copies')
            return f'{self.title} is returned, {self.available_copies} available copies'
        except Exception as e:
            logging.error(e)

    def display_book_info(self):
        return  {
            "title" : self.title,
            "author" : self.author,
            "isbn" : self.isbn,
            "publication_year" : self.publication_year,
            "available_copies" : self.available_copies
        }


book1 = Book('India that is Bharat', 'J. Sai Deepak', 1001, 2021, 7)
book2 = Book('Veer Savarkar', 'Vikram Sampath', 1002, 2021, 0)
book3 = Book('India, Bharat, Pakistan', 'J. Sai Deepak', 1003, 2023, 5)
book4 = Book('Aghora', 'Mayur Kalbag', 1004, 2020, 8)
book5 = Book('Mystics Musings', 'Sadhguru Jaggi Vasudev', 1005, 2008, 12)

print(book1.check_out())
print(book1.check_out())
print(book1.return_book())
print(book1.display_book_info())
print('-'*100)
print(book2.display_book_info())
print(book2.check_out())
```

```python
# For a ticket booking system, you have to design the "Ticket" class with OOP
# principles in mind. The “Ticket” class should have the following attributes:
# a. ticket_id: Represents the unique identifier for the ticket.
# b. event_name: Represents the name of the event.
# c. event_date: Represents the date of the event.
# d. venue: Represents the venue of the event.
# e. seat_number: Represents the seat number associated with the ticket.
# f. price: Represents the price of the ticket.
# g. is_reserved: Represents the reservation status of the ticket.
# The class also includes the following methods:
# a. reserve_ticket(self): Marks the ticket as reserved if it is not already reserved.
# b. cancel_reservation(self): Cancels the reservation of the ticket if it is already
# reserved.
# c. display_ticket_info(self): Displays the information about the ticket, including its
# attributes and reservation status.

import logging

logging.basicConfig(filename='ticket.log', format='%(levelname)s : %(message)s', level=logging.INFO)


class Ticket:

    def __init__(self, ticket_id, event_name, event_date, venue, seat_number, price, is_reserved):
        self.ticket_id = ticket_id
        self.event_name = event_name
        self.event_date = event_date
        self.venue = venue
        self.seat_number = seat_number
        self.price = price
        self.is_reserved = is_reserved

    def reserve_ticket(self):
        try:
            if not self.is_reserved:
                self.is_reserved = 1
                logging.info(f'Ticket booked for {self.event_name}, {self.event_date}, seat number {self.seat_number}')
                return f'Ticket booked for {self.event_name}, {self.event_date}. seat number {self.seat_number}'
            else:
                logging.info(f'Ticket not available for {self.event_name}, {self.event_date}')
                return f'Ticket not available for {self.event_name}, {self.event_date}'
        except Exception as e:
            logging.error(e)

    def cancel_reservation(self, isCancel):
        try:
            if isCancel and self.is_reserved:
                self.is_reserved = 0
                logging.info(f'Reservation cancelled for {self.event_name} {self.event_date}, seat number {self.seat_number}')
                return  f'Reservation cancelled for {self.event_name} {self.event_date}, seat number {self.seat_number}'
            else:
                logging.info(f'Reservation already cancelled for {self.event_name} {self.event_date}, seat number {self.seat_number}')
                return f'Reservation already cancelled for {self.event_name} {self.event_date}, seat number {self.seat_number}'
        except Exception as e:
            logging.error(e)


    def display_ticket_info(self):
            return {
                "ticket_id" : self.ticket_id,
                "event_name" : self.event_name,
                "event_date" : self.event_date,
                "venue": self.venue,
                "seat_number" : self.seat_number,
                "price" : self.price,
                "is_reserved": self.is_reserved
            }


seat1 = Ticket(101, 'Pune-Mumbai Intercity', '30-07-2023', 'Pune', 'A101', '1200', 0)
seat2 = Ticket(102, 'Banglore-Hydrabad Express', '30-07-2023', 'Pune', 'S15-9', '2200', 1)
seat3 = Ticket(103, 'Mumbai-Delhi Duranto Express', '30-07-2023', 'Pune', 'B101', '3200', 1)
seat4 = Ticket(104, 'Mumbai-Ahmadabad Express', '30-07-2023', 'Pune', 'S9-10', '1500', 0)

print(seat1.display_ticket_info())
print(seat1.reserve_ticket())
print(seat1.display_ticket_info())
print('-'*100)
print(seat2.cancel_reservation(1))
print('-'*100)
print(seat4.reserve_ticket())
print('-'*100)
print(seat2.cancel_reservation(1))

```

```python
# You are creating a shopping cart for an e-commerce website. Using OOP to model
# the "ShoppingCart" functionality the class should contain following attributes and
# methods:
# a. items: Represents the list of items in the shopping cart.
# The class also includes the following methods:
# a. add_item(self, item): Adds an item to the shopping cart by appending it to the
# list of items.
# b. remove_item(self, item): Removes an item from the shopping cart if it exists in
# the list.
# c. view_cart(self): Displays the items currently present in the shopping cart.
# d. clear_cart(self): Clears all items from the shopping cart by reassigning an
# empty list to the items attribute.

import logging

logging.basicConfig(filename='ShoppingCart.log', format='%(levelname)s : %(message)s', level=logging.INFO)


class ShoppingCart:
    items = []

    def add_item(self, item):
        try:
            self.items.append(item)
            logging.info(f'Item added in cart: {item}')
            return {'message': ' Item added in cart', 'item': item}
        except Exception as e:
            logging.error(e)

    def remove_item(self, itemId):
        try:
            for i in range(len(self.items)):
                if self.items[i]['id'] == itemId:
                    removedItem = self.items[i]
                    del self.items[i]
                    break
            logging.info(f'Item removed from cart {removedItem}')
            return {'message': 'Item removed from cart', 'removedItem': removedItem}
        except Exception as e:
            logging.error(e)

    def view_cart(self):
        return {'message': ' Items available in cart', 'items': self.items}

    def clear_cart(self):
        self.items.clear()
        logging.info('Shopping cart is cleared')
        return 'Shopping cart is cleared'


cart = ShoppingCart()
print('-'*100)
print(cart.add_item({
    'id': 1,
    'itemName': 'Prestige pressure cooker',
    'price': '2500'
}))

print(cart.add_item({
    'id': 2,
    'itemName': 'Cutting pad',
    'price': '320'
}))

print(cart.add_item({
    'id': 3,
    'itemName': 'Swiss Knife set',
    'price': '1600'
}))

print(cart.add_item({
    'id': 4,
    'itemName': 'Juicer',
    'price': '3500'
}))
print('-'*100)
print(cart.view_cart())
print('-'*100)
print(cart.remove_item(3))
print('-'*100)
print(cart.view_cart())
print('-'*100)
print(cart.clear_cart())
print('-'*100)
```

```python
#  Imagine a school management system. You have to design the "Student" class using
# OOP concepts.The “Student” class has the following attributes:
# a. name: Represents the name of the student.
# b. age: Represents the age of the student.
# c. grade: Represents the grade or class of the student.
# d. student_id: Represents the unique identifier for the student.
# e. attendance: Represents the attendance record of the student.
# The class should also include the following methods:
# a. update_attendance(self, date, status): Updates the attendance record of the
# student for a given date with the provided status (e.g., present or absent).
# b. get_attendance(self): Returns the attendance record of the student.
# c. get_average_attendance(self): Calculates and returns the average
# attendance percentage of the student based on their attendance record.


import logging

logging.basicConfig(filename='Student.log', format='%(levelname)s : %(message)s', level=logging.INFO)


class Student:

    def __init__(self, name, age, grade, student_id, attendance):
        self.name = name
        self.age = age
        self.grade = grade
        self.student_id = student_id
        self.attendance = attendance

    def update_attendance(self, attendance):
        try:
            for i in range(len(self.attendance)):
                if attendance['attendanceDate'] == self.attendance[i]['attendanceDate']:
                    self.attendance[i] = attendance
                    break
            logging.info(f'Attendance added for {self.name}')
            return {'message': f'Attendance updated for {self.name}', 'Attendance': self.attendance}
        except Exception as e:
            logging.error(e)

    def get_attendance(self):
        try:
            logging.info(f'Attendance report of {self.name}: {self.attendance}')
            return f'Attendance report of {self.name}: {self.attendance}'
        except Exception as e:
            logging.error(e)

    def get_average_attendance(self):
        listOfPresent = [x for x in self.attendance if x.get('attendance') == 'Present']
        listOfAbsent = [x for x in self.attendance if x.get('attendance') == 'Absent']
        presenteePercentage = (len(listOfPresent)/len(self.attendance)) * 100
        absenteePercentage = (len(listOfAbsent) / len(self.attendance)) * 100
        return {'presenteePercentage' : presenteePercentage, 'absenteePercentage': absenteePercentage}


student1 = Student('Student 1', 18, 12, 101, [
    {'attendanceDate': '01-June-2023', 'attendance': 'Present'},
    {'attendanceDate': '02-June-2023', 'attendance': 'Absent'},
    {'attendanceDate': '03-June-2023', 'attendance': 'Present'},
    {'attendanceDate': '04-June-2023', 'attendance': 'Absent'},
    {'attendanceDate': '05-June-2023', 'attendance': 'Absent'}
])
student2 = Student('Student 2', 18, 12, 102, [
    {'attendanceDate': '01-June-2023', 'attendance': 'Present'},
    {'attendanceDate': '02-June-2023', 'attendance': 'Present'},
    {'attendanceDate': '03-June-2023', 'attendance': 'Present'},
    {'attendanceDate': '04-June-2023', 'attendance': 'Present'},
    {'attendanceDate': '05-June-2023', 'attendance': 'Absent'}
])
student3 = Student('Student 3', 18, 12, 103, [
    {'attendanceDate': '01-June-2023', 'attendance': 'Absent'},
    {'attendanceDate': '02-June-2023', 'attendance': 'Absent'},
    {'attendanceDate': '03-June-2023', 'attendance': 'Absent'},
    {'attendanceDate': '04-June-2023', 'attendance': 'Absent'},
    {'attendanceDate': '05-June-2023', 'attendance': 'Absent'}
])
student4 = Student('Student 4', 18, 12, 104, [
    {'attendanceDate': '01-June-2023', 'attendance': 'Present'},
    {'attendanceDate': '02-June-2023', 'attendance': 'Present'},
    {'attendanceDate': '03-June-2023', 'attendance': 'Present'},
    {'attendanceDate': '04-June-2023', 'attendance': 'Present'},
    {'attendanceDate': '05-June-2023', 'attendance': 'Present'}
])

print(student3.get_attendance())
print('-'*100)
print(student3.update_attendance({'attendanceDate': '05-June-2023', 'attendance': 'Present'}))
print('-'*100)
print(student1.get_average_attendance())
print('-'*100)
print(student2.get_average_attendance())
print('-'*100)
print(student4.get_average_attendance())
print('-'*100)
```
